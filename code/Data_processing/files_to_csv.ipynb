```python
from tqdm.notebook import tqdm
from time import time

import os
from glob import glob
import csv
import shutil
import pprint
```

```python
img_path = '/opt/ml/input/data/train/images'
csv_path = '/opt/ml/input/data/train/train_result.csv'

c = open(csv_path, 'w', encoding='utf-8')
fd = csv.writer(c)
fd.writerow(['id', 'person_id', 'gender', 'age', 'mask', 'label', 'path'])

s_list = []
for file_ in glob(img_path + '/*'):
    file_name = file_.split('/')[-1]
    list_ = file_name.split('.')[0].split('_')
    list_.append(file_name)
    s_list.append(list_)

to_sort = sorted(s_list)

for i in tqdm(to_sort):
    print(i)
    fd.writerow(i)
    
print(len(to_sort))


# pp = pprint.PrettyPrinter(depth=6)
# pp.pprint(to_sort)
```
