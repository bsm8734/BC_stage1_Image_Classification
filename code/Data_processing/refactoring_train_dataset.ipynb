```python
import os
import os.path
import csv
import shutil
```

```python
def label_images(tup, ind):
    ret = 0
    if tup[1] == 'female': ret += 3
    age = int(tup[3])
    if age >= 30 and age < 60: ret+=1
    elif age >= 60: ret += 2
    ret += int(ind)
    return ret
```

```python
def change_path(src, dst):
    os.rename(src, dst)
```

```python
def exist_path(csv_line, path):
    appendix_ind = [0, 0, 0, 0, 0, 6, 12] # 정상, 비정상, 미착용
    appendix_list = ['/mask1', '/mask2', '/mask3', '/mask4', '/mask5', '/incorrect_mask', '/normal']
    ext_list = ['.jpg', '.png', '.jpeg']
    
    for i in range(0, 7):
        dst = ""
        for ext in ext_list:
            if os.path.exists(path + appendix_list[i] + ext):
                dst_path = path + appendix_list[i] + ext
                break

        ans = label_images(csv_line, appendix_ind[i])
        src_path = dst_path
        if i == 5:
            dst_path = '_'.join(dst_path.split('.')[0].split('_')[:-1]) + '_' + str(ans) + '.' + dst_path.split('.')[1]
        else:
            dst_path = dst_path.split('.')[0] + '_' + str(ans) + '.' + dst_path.split('.')[1]
        print(src_path, dst_path)
#         change_path(src_path, dst_path)
            
        
        
#         # 실제 파일의 위치
#         path_dir = src_img_path + folder_path + '/' + file_name
        
    
```

```python
def main():
    abs_path = '/opt/ml/input/data/train'
    csv_path = abs_path + '/train.csv'
    img_path = abs_path + '/images/'

    src_csv = open(csv_path, 'r', encoding='utf-8')
    src_fd = csv.reader(src_csv)

    for line in src_fd:
        folder_name = line[-1]
        if folder_name == 'path': continue
        # print(img_path + folder_name) # /opt/ml/input/data/train/images/000001_female_Asian_45
        exist_path(line, img_path + folder_name)

main()
# dst_csv_path = cur_path + '/train/train.csv'
# dst_img_path = cur_path + '/train/images/'

# dst_csv = open(dst_csv_path, 'w', encoding='utf-8')
# dst_fd = csv.writer(dst_csv)
# dst_fd.writerow(['id', 'gender', 'age', 'new_path', 'label'])

# appendix_ind = [0, 0, 0, 0, 0, 6, 12] # 정상, 비정상, 미착용

# for line in src_fd:
#     folder_path = line[-1]
#     if folder_path == "path": continue
#     print(img_path + folder_path)
        
```
